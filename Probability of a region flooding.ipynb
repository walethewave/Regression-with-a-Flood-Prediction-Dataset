```python
import pandas as pd

# Load the data
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

# Display the first few rows of the datasets
print(train_data.head())
print(test_data.head())
```

```python
# Summary statistics
print(train_data.describe())
print(test_data.describe())

# Check for missing values
print(train_data.isnull().sum())
print(test_data.isnull().sum())
```

```python
# Inspect column names
print(train_data.columns)
print(test_data.columns)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

```

```python

# Separate features and target variable
X_train = train_data.drop(['id', 'FloodProbability'], axis=1)
y_train = train_data['FloodProbability']

# Perform train-validation split since test set doesn't have target variable
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(test_data.drop(['id'], axis=1))  # Scaling the test data without target
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred_train = model.predict(X_train_scaled)
y_pred_val = model.predict(X_val_scaled)

# Evaluate the Linear Regression model
train_mse = mean_squared_error(y_train, y_pred_train)
val_mse = mean_squared_error(y_val, y_pred_val)
train_r2 = r2_score(y_train, y_pred_train)
val_r2 = r2_score(y_val, y_pred_val)

print(f'Train MSE: {train_mse}, Train R2: {train_r2}')
print(f'Validation MSE: {val_mse}, Validation R2: {val_r2}')
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train)

# Make predictions
y_pred_train_rf = rf_model.predict(X_train_scaled)
y_pred_val_rf = rf_model.predict(X_val_scaled)

# Evaluate the Random Forest model
train_mse_rf = mean_squared_error(y_train, y_pred_train_rf)
val_mse_rf = mean_squared_error(y_val, y_pred_val_rf)
train_r2_rf = r2_score(y_train, y_pred_train_rf)
val_r2_rf = r2_score(y_val, y_pred_val_rf)

print(f'Train MSE (RF): {train_mse_rf}, Train R2 (RF): {train_r2_rf}')
print(f'Validation MSE (RF): {val_mse_rf}, Validation R2 (RF): {val_r2_rf}')
```
